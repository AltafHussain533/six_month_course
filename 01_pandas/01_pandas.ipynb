```python
# 2- Filling missing values of age column by mean of age
df['age'] = df['age'].fillna(df['age'].mean())
# df['age'].fillna(df['age'].median(), inplace=True)  
```

```python
df['age'].mode() [0] #  correct way to fill missing values of age column by mode of age column because mode is not a single value it is a series of values
```

```python
# Fill embark and embarked_town columns by mode of embark and embarked_town columns
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode())   
```

```python
df.isnull().sum()
```

```python
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
```

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```
