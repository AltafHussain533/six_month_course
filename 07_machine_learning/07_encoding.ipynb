# Feature Encoding

Feature encoding is the process of transforming `categorical features` into `numeric features`. This is necessary because machine learning algorithms can only handle numeric features. There are many different ways to encode categorical features, and each method has its own advantages and disadvantages. In this notebook, we will explore some of the most popular methods for encoding categorical features, such as:

- Label encoding
- Ordinal encoding
- One-hot encoding
- Binary encoding

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# data load
df = sns.load_dataset('tips')
df.head()
```

## 1. **Label encoding**
This technique assigns a unique integer to each category sequentially based on alphabetical order or appearance.Mechanism: Maps N categories to integers from 0 to N-1.
Example: ["Low", "Medium", "High"]-- [0, 1, 2].
When to use: Target variables (y) only. Do not use it for input features (X) because algorithms will mistakenly assume math relationships (e.g., that "High" (2) is twice as much as "Medium" (1)).

```python
df['time'].value_counts()
```

```python
# let's encode the time in labelencoder with sklearn
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
le = LabelEncoder()
df['encoded_time'] = le.fit_transform(df['time'])
df.head()
```

```python
df['time'].value_counts()
```

```python
df['encoded_time'].value_counts()
```

## 2. **Ordinal Encoding**   
This is a specialized version of label encoding where you manually define the integer order to preserve a known ranking or hierarchy.\
**Mechanism:** Explicitly maps categories to specific numeric scales.Example: {"Cold": 0, "Warm": 1, "Hot": 2}.\
**When to use:** Input features (X) that have a strict, inherent ranking (e.g., education level, survey responses like Likert scales, or shirt sizes).

```python
df['day'].value_counts()
```

```python
# ordinal encoding the day column using specific order
oe = OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])
df['encoded_day'] = oe.fit_transform(df[['day']])
df.head(60)
```

```python
df['encoded_day'].value_counts(ascending=True)
```

```python
# one hot encoding on day column
ohe = OneHotEncoder()
ohe.fit_transform(df[['sex']]).toarray()
```

```python
# example of one hot encoding
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# example of one hot encoding
titanic = sns.load_dataset('titanic')

onehot_encoder = OneHotEncoder(sparse=False)
embarked_onehot = onehot_encoder.fit_transform(titanic[['embarked']])
embarked_onehot_df = pd.DataFrame(embarked_onehot, columns=onehot_encoder.get_feature_names_out(['embarked']))
titanic = pd.concat([titanic.reset_index(drop=True), embarked_onehot_df.reset_index(drop=True)], axis=1)
titanic.head()
```
